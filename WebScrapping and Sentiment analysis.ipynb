```python
import string

# filtere
stpw = [element.lower() for element in stpw]

# Function to remove stop words from text
def remove_stop_words(text, stpw):
    tokens = word_tokenize(text)  # Tokenize the text
    filtered_tokens = [word for word in tokens if word not in stpw]  # Remove stop words
    filtered_tokens = [word.translate(str.maketrans('', '', string.punctuation)) for word in filtered_tokens]

    return ' '.join(filtered_tokens)  # Join tokens back into a string

# Remove stop words from columns of the DataFrame
#for column in df.columns:
extract['content_x'] = extract['content'].apply(lambda x: remove_stop_words(x, stpw))

# Save the modified DataFrame back to the Excel file
#df.to_excel('output.xlsx', index=False)
```

```python
import os
sentment_dir = "/content/drive/MyDrive/project/drive-download-20240202T232102Z-001/MasterDictionary"
pos=set()
neg=set()

for files in os.listdir(sentment_dir):
  if files =='positive-words.txt':
    with open(os.path.join(sentment_dir,files),'r',encoding='ISO-8859-1') as f:
      pos.update(f.read().splitlines())
  else:
    with open(os.path.join(sentment_dir,files),'r',encoding='ISO-8859-1') as f:
      neg.update(f.read().splitlines())
```

```python
positive_words = []
Negative_words =[]
positive_score = []
negative_score = []
polarity_score = []
subjectivity_score = []

#iterate through the list of docs
for i in range(len(token_sets_by_row)):
  positive_words.append([word for word in token_sets_by_row[i] if word.lower() in pos])
  Negative_words.append([word for word in token_sets_by_row[i] if word.lower() in neg])
  positive_score.append(len(positive_words[i]))
  negative_score.append(len(Negative_words[i]))
  polarity_score.append((positive_score[i] - negative_score[i]) / ((positive_score[i] + negative_score[i]) + 0.000001))
  subjectivity_score.append((positive_score[i] + negative_score[i]) / ((len(token_sets_by_row[i])) + 0.000001))
```

```python
 print("Positive Words:", positive_words)
 print("Negative Words:", Negative_words)
print("Positive Score:", positive_score)
 print("Negative Score:", negative_score)
 print("Polarity Score:", polarity_score)
 print("Subjectivity Score:", subjectivity_score)
```

```python
# prompt: write code to calculate average sentence length of the each row in column content and percentage of complex also  avg_syllable_word_count = syllable_count / len(syllable_words), Fog_Index = 0.4 * (avg_sentence_len + Percent_Complex_words)

def count_sentences(text):
  sentences = nltk.sent_tokenize(text)
  return len(sentences)

def count_complex_sentences(text):
  sentences = nltk.sent_tokenize(text)
  complex_sentences = 0
  for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    if len(words) > 10:
      complex_sentences += 1
  return complex_sentences

def count_syllables(word):
  count = 0
  vowels = "aeiouy"
  word = word.lower()
  if word[0] in vowels:
    count += 1
  for i in range(1, len(word)):
    if word[i] in vowels and word[i - 1] not in vowels:
      count += 1
  if word.endswith("e"):
    count -= 1
  if count == 0:
    count += 1
  return count

def count_complex_words(text):
  words = nltk.word_tokenize(text)
  complex_words = 0
  for word in words:
    if count_syllables(word) > 2:
      complex_words += 1
  return complex_words

def calculate_avg_sentence_length(text):
  sentences = nltk.sent_tokenize(text)
  total_words = 0
  for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    total_words += len(words)
  avg_sentence_length = total_words / len(sentences)
  return avg_sentence_length

def calculate_percent_complex_sentences(text):
  sentences = nltk.sent_tokenize(text)
  complex_sentences = 0
  for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    if len(words) > 10:
      complex_sentences += 1
  percent_complex_sentences = complex_sentences / len(sentences) * 100
  return percent_complex_sentences

def calculate_avg_syllable_word_count(text):
  words = nltk.word_tokenize(text)
  syllable_words = []
  for word in words:
    if count_syllables(word) > 0:
      syllable_words.append(word)
  syllable_count = 0
  for word in syllable_words:
    syllable_count += count_syllables(word)
  avg_syllable_word_count = syllable_count / len(syllable_words)
  return avg_syllable_word_count

def calculate_fog_index(text):
  avg_sentence_length = calculate_avg_sentence_length(text)
  percent_complex_sentences = calculate_percent_complex_sentences(text)
  fog_index = 0.4 * (avg_sentence_length + percent_complex_sentences)
  return fog_index

# Calculate the required metrics for each row in the "content" column
extract['avg_sentence_length'] = extract['content'].apply(calculate_avg_sentence_length)
extract['percent_complex_sentences'] = extract['content'].apply(calculate_percent_complex_sentences)
extract['avg_syllable_word_count'] = extract['content'].apply(calculate_avg_syllable_word_count)
extract['fog_index'] = extract['content'].apply(calculate_fog_index)

```
